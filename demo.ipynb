```python
# ÂÖãÈöÜÂπ∂ÂØºÂÖ•KarpathyÁöÑnanoGPT

import os
import sys
import torch
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio

# =====================
# 1. ÂÆâË£ÖÂíåÂØºÂÖ•nanoGPT
# =====================

def setup_nanogpt():
    """ËÆæÁΩÆnanoGPTÁéØÂ¢É"""
    print("ËÆæÁΩÆnanoGPT")

    # ÂÖãÈöÜnanoGPT‰ªìÂ∫ì
    if not os.path.exists('nanoGPT'):
        print("clone")
        os.system('git clone https://github.com/karpathy/nanoGPT.git')

    # Ê∑ªÂä†Âà∞PythonË∑ØÂæÑ
    sys.path.append('./nanoGPT')

    # ÂØºÂÖ•nanoGPTÊ®°Âùó
    try:
        from model import GPTConfig, GPT
        print("nanoGPTÂØºÂÖ•ÊàêÂäü!")
        return GPTConfig, GPT
    except ImportError as e:
        print(f"error")
        return None, None

# =====================
# 2. Èü≥‰πêÊï∞ÊçÆÂ§ÑÁêÜ
# =====================

class MusicTokenizer:
    """Èü≥‰πêtokenizer"""
    def __init__(self):
        self.notes = [
            # Âü∫Êú¨Èü≥Á¨¶
            'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B',
            # È´òÂÖ´Â∫¶
            'C5', 'C#5', 'D5', 'D#5', 'E5', 'F5', 'F#5', 'G5', 'G#5', 'A5', 'A#5', 'B5',
            # ‰ΩéÂÖ´Â∫¶
            'C3', 'C#3', 'D3', 'D#3', 'E3', 'F3', 'F#3', 'G3', 'G#3', 'A3', 'A#3', 'B3',
            # ÁâπÊÆäÁ¨¶Âè∑
            'REST', 'BAR', 'START', 'END'
        ]

        self.note_to_int = {note: i for i, note in enumerate(self.notes)}
        self.int_to_note = {i: note for i, note in enumerate(self.notes)}
        self.vocab_size = len(self.notes)

    def encode(self, music_sequence):
        return [self.note_to_int[note] for note in music_sequence if note in self.note_to_int]

    def decode(self, indices):
        return [self.int_to_note[i] for i in indices if i in self.int_to_note]

def create_advanced_music_data():
    """ÂàõÂª∫Êõ¥Â§çÊùÇÁöÑÈü≥‰πêËÆ≠ÁªÉÊï∞ÊçÆ"""

    # ÂêÑÁßçÈü≥‰πêÊ®°Âºè
    patterns = {
        'scales': [
            # Â§ßË∞ÉÈü≥Èò∂
            ['C', 'D', 'E', 'F', 'G', 'A', 'B', 'C5', 'BAR'],
            ['G', 'A', 'B', 'C5', 'D5', 'E5', 'F#5', 'G5', 'BAR'],
            # Â∞èË∞ÉÈü≥Èò∂
            ['A', 'B', 'C5', 'D5', 'E5', 'F5', 'G5', 'A5', 'BAR'],
        ],
        'arpeggios': [
            # ‰∏âÂíåÂº¶ÂàÜËß£
            ['C', 'E', 'G', 'C5', 'G', 'E', 'C', 'BAR'],
            ['F', 'A', 'C5', 'F5', 'C5', 'A', 'F', 'BAR'],
            ['G', 'B', 'D5', 'G5', 'D5', 'B', 'G', 'BAR'],
        ],
        'melodies': [
            # ÁªèÂÖ∏ÊóãÂæãÁâáÊÆµ
            ['C', 'C', 'G', 'G', 'A', 'A', 'G', 'REST', 'BAR'],
            ['F', 'F', 'E', 'E', 'D', 'D', 'C', 'REST', 'BAR'],
            ['E', 'E', 'F', 'G', 'G', 'F', 'E', 'D', 'BAR'],
        ],
        'rhythmic': [
            # ËäÇÂ•èÊ®°Âºè
            ['C', 'REST', 'C', 'REST', 'E', 'REST', 'G', 'REST', 'BAR'],
            ['G', 'G', 'REST', 'F', 'F', 'REST', 'E', 'D', 'BAR'],
        ]
    }

    # ÁîüÊàêËÆ≠ÁªÉÊï∞ÊçÆ
    music_data = ['START']

    for _ in range(100):  # ÁîüÊàê100‰∏™Èü≥‰πêÁâáÊÆµ
        category = np.random.choice(list(patterns.keys()))
        pattern_list = patterns[category]
        pattern = pattern_list[np.random.randint(len(pattern_list))]
        music_data.extend(pattern)

    music_data.append('END')
    return music_data

# =====================
# 3. ‰ΩøÁî®ÂéüÁâànanoGPTËÆ≠ÁªÉ
# =====================

def train_music_with_nanogpt():
    """ÂéüÁâànanoGPTËÆ≠ÁªÉÈü≥‰πêÊ®°Âûã"""

    # ËÆæÁΩÆnanoGPT
    GPTConfig, GPT = setup_nanogpt()
    if GPTConfig is None:
        return None, None, None

    print("ÂáÜÂ§áÊï∞ÊçÆ")

    # ÂáÜÂ§áÊï∞ÊçÆ
    tokenizer = MusicTokenizer()
    music_data = create_advanced_music_data()

    print(f"Èü≥‰πêÊï∞ÊçÆÊ†∑Êú¨: {music_data[:30]}")
    print(f"ÊÄªÈïøÂ∫¶: {len(music_data)}")

    # ÁºñÁ†ÅÊï∞ÊçÆ
    encoded_data = tokenizer.encode(music_data)
    data = torch.tensor(encoded_data, dtype=torch.long)

    # ÈÖçÁΩÆnanoGPTÊ®°Âûã
    config = GPTConfig(
        block_size=128,         # Â∫èÂàóÈïøÂ∫¶
        vocab_size=tokenizer.vocab_size,  # ËØçÊ±áË°®Â§ßÂ∞è
        n_layer=6,              # TransformerÂ±ÇÊï∞
        n_head=6,               # Ê≥®ÊÑèÂäõÂ§¥Êï∞
        n_embd=192,             # ÂµåÂÖ•Áª¥Â∫¶
        dropout=0.1,            # DropoutÁéá
        bias=False,             # ÊòØÂê¶‰ΩøÁî®bias
    )

    # ÂàõÂª∫Ê®°Âûã
    model = GPT(config)
    print(f"Ê®°ÂûãÂèÇÊï∞Êï∞Èáè: {sum(p.numel() for p in model.parameters())/1e6:.2f}M")

    # ËÆ≠ÁªÉËÆæÁΩÆ
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model = model.to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4)

    # ËÆ≠ÁªÉÂæ™ÁéØ
    model.train()
    losses = []

    print(f"\nÂºÄÂßãËÆ≠ÁªÉ ({device})...")

    for iter in range(1000):  # ËÆ≠ÁªÉ1000Ê≠•
        # ÂáÜÂ§áÊâπÊ¨°Êï∞ÊçÆ
        batch_size = 8
        block_size = config.block_size

        if len(data) <= block_size:
            continue

        # ÈöèÊú∫ÈááÊ†∑
        ix = torch.randint(len(data) - block_size, (batch_size,))
        x = torch.stack([data[i:i+block_size] for i in ix]).to(device)
        y = torch.stack([data[i+1:i+block_size+1] for i in ix]).to(device)

        # ÂâçÂêë‰º†Êí≠
        logits, loss = model(x, y)

        # ÂèçÂêë‰º†Êí≠
        optimizer.zero_grad(set_to_none=True)
        loss.backward()
        optimizer.step()

        losses.append(loss.item())

        if iter % 200 == 0:
            print(f"Ê≠•È™§ {iter:4d}, ÊçüÂ§±: {loss.item():.4f}")

    return model, tokenizer, losses

def generate_music_with_nanogpt(model, tokenizer, seed=['START', 'C', 'E'], max_new_tokens=50):
    """‰ΩøÁî®ËÆ≠ÁªÉÂ•ΩÁöÑnanoGPTÁîüÊàêÈü≥‰πê"""

    model.eval()
    device = next(model.parameters()).device

    # ÁºñÁ†ÅÁßçÂ≠ê
    seed_encoded = tokenizer.encode(seed)
    context = torch.tensor(seed_encoded, dtype=torch.long).unsqueeze(0).to(device)

    print(f"ÁßçÂ≠ê: {seed}")

    # ÁîüÊàêÈü≥‰πê
    with torch.no_grad():
        generated = model.generate(context, max_new_tokens, temperature=0.8, top_k=20)
        generated_notes = tokenizer.decode(generated[0].cpu().tolist())

    # Ê∏ÖÁêÜÁîüÊàêÁªìÊûú
    if 'END' in generated_notes:
        end_idx = generated_notes.index('END')
        generated_notes = generated_notes[:end_idx]

    print(f"ÁîüÊàêÁöÑÈü≥‰πê: {generated_notes}")
    return generated_notes

# =====================
# 4. Èü≥È¢ëÁîüÊàê (‰∏é‰πãÂâçÁõ∏Âêå)
# =====================

def note_to_frequency(note):
    """Èü≥Á¨¶Âà∞È¢ëÁéáÁöÑÊò†Â∞Ñ"""
    base_freq = {
        'C': 261.63, 'C#': 277.18, 'D': 293.66, 'D#': 311.13,
        'E': 329.63, 'F': 349.23, 'F#': 369.99, 'G': 392.00,
        'G#': 415.30, 'A': 440.00, 'A#': 466.16, 'B': 493.88
    }

    if note in ['REST', 'BAR', 'START', 'END']:
        return 0.0

    # Â§ÑÁêÜÂÖ´Â∫¶
    octave_mult = 1.0
    clean_note = note

    if note.endswith('3'):
        octave_mult = 0.5
        clean_note = note[:-1]
    elif note.endswith('5'):
        octave_mult = 2.0
        clean_note = note[:-1]

    return base_freq.get(clean_note, 0.0) * octave_mult

def generate_audio_from_notes(notes, duration=0.4, sample_rate=22050):
    """‰ªéÈü≥Á¨¶ÁîüÊàêÈü≥È¢ë"""
    audio = []
    t = np.linspace(0, duration, int(sample_rate * duration), False)

    for note in notes:
        freq = note_to_frequency(note)
        if freq > 0:
            # ÁîüÊàêÈü≥Á¨¶
            wave = np.sin(2 * np.pi * freq * t) * 0.3
            # Ê∑ªÂä†ÂåÖÁªú
            envelope = np.exp(-t * 2) + 0.1
            wave = wave * envelope
        else:
            wave = np.zeros_like(t)

        audio.extend(wave)

    return np.array(audio), sample_rate

# =====================
# 5. ‰∏ªÊºîÁ§∫ÂáΩÊï∞
# =====================

def main_demo():
    """‰∏ªÊºîÁ§∫ÂáΩÊï∞ - ‰ΩøÁî®ÂéüÁâànanoGPT"""
    print("=" * 60)
    print("nanoGPT (ÂéüÁâà) Èü≥‰πêÁîüÊàêÊºîÁ§∫")
    print("=" * 60)

    # ËÆ≠ÁªÉÊ®°Âûã
    model, tokenizer, losses = train_music_with_nanogpt()

    if model is None:
        print("Ê®°ÂûãËÆ≠ÁªÉÂ§±Ë¥•ÔºåËØ∑Ê£ÄÊü•nanoGPTÂÆâË£Ö")
        return

    # ÁªòÂà∂ËÆ≠ÁªÉÊõ≤Á∫ø
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(losses)
    plt.title('ËÆ≠ÁªÉÊçüÂ§±Êõ≤Á∫ø')
    plt.xlabel('ËÆ≠ÁªÉÊ≠•È™§')
    plt.ylabel('ÊçüÂ§±ÂÄº')
    plt.grid(True)

    # ÁîüÊàêÈü≥‰πê
    print("\n" + "=" * 50)
    print("üéµ ‰ΩøÁî®ËÆ≠ÁªÉÂ•ΩÁöÑÊ®°ÂûãÁîüÊàêÈü≥‰πê...")
    print("=" * 50)

    seeds = [
        ['START', 'C', 'E', 'G'],
        ['START', 'F', 'A', 'C5'],
        ['START', 'G', 'B', 'D5'],
        ['START', 'C'],
    ]

    all_generated = []

    for i, seed in enumerate(seeds):
        print(f"\n--- ÁîüÊàêÂ∫èÂàó {i+1} ---")
        generated = generate_music_with_nanogpt(model, tokenizer, seed, max_new_tokens=20)
        # ËøáÊª§ÊéâÁâπÊÆäÁ¨¶Âè∑
        filtered = [note for note in generated if note not in ['START', 'END', 'BAR']]
        all_generated.extend(filtered[:12])  # ÊØè‰∏™Â∫èÂàóÂèñ12‰∏™Èü≥Á¨¶

    # ÂèØËßÜÂåñÁîüÊàêÁöÑÈü≥‰πê
    plt.subplot(1, 2, 2)
    if all_generated:
        note_indices = [tokenizer.note_to_int.get(note, 0) for note in all_generated[:32]]
        plt.plot(note_indices, 'o-', markersize=4)
        plt.title('ÁîüÊàêÁöÑÈü≥‰πêÂ∫èÂàó')
        plt.xlabel('Êó∂Èó¥Ê≠•')
        plt.ylabel('Èü≥Á¨¶ÁºñÂè∑')
        plt.grid(True)

    plt.tight_layout()
    plt.show()

    # ÁîüÊàêÈü≥È¢ë
    if all_generated:
        print(f"\nÁîüÊàêÈü≥È¢ë (ÂÖ±{len(all_generated)}‰∏™Èü≥Á¨¶)...")
        audio_data, sample_rate = generate_audio_from_notes(all_generated[:24])

        print(f"\nÁîüÊàêÁöÑÈü≥Á¨¶Â∫èÂàó:")
        print(" -> ".join(all_generated[:24]))

        return audio_data, sample_rate, all_generated[:24]
    else:
        print("Ê≤°ÊúâÁîüÊàêÊúâÊïàÁöÑÈü≥‰πêÂ∫èÂàó")
        return None, None, []

# =====================
# 6. ËøêË°åÊºîÁ§∫
# =====================

if __name__ == "__main__":
    # ËøêË°åÊºîÁ§∫
    result = main_demo()

    if result[0] is not None:
        audio, sr, notes = result
        print(f"\nÊí≠ÊîæÁîüÊàêÁöÑÈü≥‰πê...")
        display(Audio(audio, rate=sr))


    print("ÂÆåÊàê")
```
