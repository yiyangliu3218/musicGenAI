```python
# 克隆并导入Karpathy的nanoGPT

import os
import sys
import torch
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio

# =====================
# 1. 安装和导入nanoGPT
# =====================

def setup_nanogpt():
    """设置nanoGPT环境"""
    print("设置nanoGPT")

    # 克隆nanoGPT仓库
    if not os.path.exists('nanoGPT'):
        print("clone")
        os.system('git clone https://github.com/karpathy/nanoGPT.git')

    # 添加到Python路径
    sys.path.append('./nanoGPT')

    # 导入nanoGPT模块
    try:
        from model import GPTConfig, GPT
        print("nanoGPT导入成功!")
        return GPTConfig, GPT
    except ImportError as e:
        print(f"error")
        return None, None

# =====================
# 2. 音乐数据处理
# =====================

class MusicTokenizer:
    """音乐tokenizer"""
    def __init__(self):
        self.notes = [
            # 基本音符
            'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B',
            # 高八度
            'C5', 'C#5', 'D5', 'D#5', 'E5', 'F5', 'F#5', 'G5', 'G#5', 'A5', 'A#5', 'B5',
            # 低八度
            'C3', 'C#3', 'D3', 'D#3', 'E3', 'F3', 'F#3', 'G3', 'G#3', 'A3', 'A#3', 'B3',
            # 特殊符号
            'REST', 'BAR', 'START', 'END'
        ]

        self.note_to_int = {note: i for i, note in enumerate(self.notes)}
        self.int_to_note = {i: note for i, note in enumerate(self.notes)}
        self.vocab_size = len(self.notes)

    def encode(self, music_sequence):
        return [self.note_to_int[note] for note in music_sequence if note in self.note_to_int]

    def decode(self, indices):
        return [self.int_to_note[i] for i in indices if i in self.int_to_note]

def create_advanced_music_data():
    """创建更复杂的音乐训练数据"""

    # 各种音乐模式
    patterns = {
        'scales': [
            # 大调音阶
            ['C', 'D', 'E', 'F', 'G', 'A', 'B', 'C5', 'BAR'],
            ['G', 'A', 'B', 'C5', 'D5', 'E5', 'F#5', 'G5', 'BAR'],
            # 小调音阶
            ['A', 'B', 'C5', 'D5', 'E5', 'F5', 'G5', 'A5', 'BAR'],
        ],
        'arpeggios': [
            # 三和弦分解
            ['C', 'E', 'G', 'C5', 'G', 'E', 'C', 'BAR'],
            ['F', 'A', 'C5', 'F5', 'C5', 'A', 'F', 'BAR'],
            ['G', 'B', 'D5', 'G5', 'D5', 'B', 'G', 'BAR'],
        ],
        'melodies': [
            # 经典旋律片段
            ['C', 'C', 'G', 'G', 'A', 'A', 'G', 'REST', 'BAR'],
            ['F', 'F', 'E', 'E', 'D', 'D', 'C', 'REST', 'BAR'],
            ['E', 'E', 'F', 'G', 'G', 'F', 'E', 'D', 'BAR'],
        ],
        'rhythmic': [
            # 节奏模式
            ['C', 'REST', 'C', 'REST', 'E', 'REST', 'G', 'REST', 'BAR'],
            ['G', 'G', 'REST', 'F', 'F', 'REST', 'E', 'D', 'BAR'],
        ]
    }

    # 生成训练数据
    music_data = ['START']

    for _ in range(100):  # 生成100个音乐片段
        category = np.random.choice(list(patterns.keys()))
        pattern_list = patterns[category]
        pattern = pattern_list[np.random.randint(len(pattern_list))]
        music_data.extend(pattern)

    music_data.append('END')
    return music_data

# =====================
# 3. 使用原版nanoGPT训练
# =====================

def train_music_with_nanogpt():
    """原版nanoGPT训练音乐模型"""

    # 设置nanoGPT
    GPTConfig, GPT = setup_nanogpt()
    if GPTConfig is None:
        return None, None, None

    print("准备数据")

    # 准备数据
    tokenizer = MusicTokenizer()
    music_data = create_advanced_music_data()

    print(f"音乐数据样本: {music_data[:30]}")
    print(f"总长度: {len(music_data)}")

    # 编码数据
    encoded_data = tokenizer.encode(music_data)
    data = torch.tensor(encoded_data, dtype=torch.long)

    # 配置nanoGPT模型
    config = GPTConfig(
        block_size=128,         # 序列长度
        vocab_size=tokenizer.vocab_size,  # 词汇表大小
        n_layer=6,              # Transformer层数
        n_head=6,               # 注意力头数
        n_embd=192,             # 嵌入维度
        dropout=0.1,            # Dropout率
        bias=False,             # 是否使用bias
    )

    # 创建模型
    model = GPT(config)
    print(f"模型参数数量: {sum(p.numel() for p in model.parameters())/1e6:.2f}M")

    # 训练设置
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model = model.to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4)

    # 训练循环
    model.train()
    losses = []

    print(f"\n开始训练 ({device})...")

    for iter in range(1000):  # 训练1000步
        # 准备批次数据
        batch_size = 8
        block_size = config.block_size

        if len(data) <= block_size:
            continue

        # 随机采样
        ix = torch.randint(len(data) - block_size, (batch_size,))
        x = torch.stack([data[i:i+block_size] for i in ix]).to(device)
        y = torch.stack([data[i+1:i+block_size+1] for i in ix]).to(device)

        # 前向传播
        logits, loss = model(x, y)

        # 反向传播
        optimizer.zero_grad(set_to_none=True)
        loss.backward()
        optimizer.step()

        losses.append(loss.item())

        if iter % 200 == 0:
            print(f"步骤 {iter:4d}, 损失: {loss.item():.4f}")

    return model, tokenizer, losses

def generate_music_with_nanogpt(model, tokenizer, seed=['START', 'C', 'E'], max_new_tokens=50):
    """使用训练好的nanoGPT生成音乐"""

    model.eval()
    device = next(model.parameters()).device

    # 编码种子
    seed_encoded = tokenizer.encode(seed)
    context = torch.tensor(seed_encoded, dtype=torch.long).unsqueeze(0).to(device)

    print(f"种子: {seed}")

    # 生成音乐
    with torch.no_grad():
        generated = model.generate(context, max_new_tokens, temperature=0.8, top_k=20)
        generated_notes = tokenizer.decode(generated[0].cpu().tolist())

    # 清理生成结果
    if 'END' in generated_notes:
        end_idx = generated_notes.index('END')
        generated_notes = generated_notes[:end_idx]

    print(f"生成的音乐: {generated_notes}")
    return generated_notes

# =====================
# 4. 音频生成 (与之前相同)
# =====================

def note_to_frequency(note):
    """音符到频率的映射"""
    base_freq = {
        'C': 261.63, 'C#': 277.18, 'D': 293.66, 'D#': 311.13,
        'E': 329.63, 'F': 349.23, 'F#': 369.99, 'G': 392.00,
        'G#': 415.30, 'A': 440.00, 'A#': 466.16, 'B': 493.88
    }

    if note in ['REST', 'BAR', 'START', 'END']:
        return 0.0

    # 处理八度
    octave_mult = 1.0
    clean_note = note

    if note.endswith('3'):
        octave_mult = 0.5
        clean_note = note[:-1]
    elif note.endswith('5'):
        octave_mult = 2.0
        clean_note = note[:-1]

    return base_freq.get(clean_note, 0.0) * octave_mult

def generate_audio_from_notes(notes, duration=0.4, sample_rate=22050):
    """从音符生成音频"""
    audio = []
    t = np.linspace(0, duration, int(sample_rate * duration), False)

    for note in notes:
        freq = note_to_frequency(note)
        if freq > 0:
            # 生成音符
            wave = np.sin(2 * np.pi * freq * t) * 0.3
            # 添加包络
            envelope = np.exp(-t * 2) + 0.1
            wave = wave * envelope
        else:
            wave = np.zeros_like(t)

        audio.extend(wave)

    return np.array(audio), sample_rate

# =====================
# 5. 主演示函数
# =====================

def main_demo():
    """主演示函数 - 使用原版nanoGPT"""
    print("=" * 60)
    print("nanoGPT (原版) 音乐生成演示")
    print("=" * 60)

    # 训练模型
    model, tokenizer, losses = train_music_with_nanogpt()

    if model is None:
        print("模型训练失败，请检查nanoGPT安装")
        return

    # 绘制训练曲线
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(losses)
    plt.title('训练损失曲线')
    plt.xlabel('训练步骤')
    plt.ylabel('损失值')
    plt.grid(True)

    # 生成音乐
    print("\n" + "=" * 50)
    print("🎵 使用训练好的模型生成音乐...")
    print("=" * 50)

    seeds = [
        ['START', 'C', 'E', 'G'],
        ['START', 'F', 'A', 'C5'],
        ['START', 'G', 'B', 'D5'],
        ['START', 'C'],
    ]

    all_generated = []

    for i, seed in enumerate(seeds):
        print(f"\n--- 生成序列 {i+1} ---")
        generated = generate_music_with_nanogpt(model, tokenizer, seed, max_new_tokens=20)
        # 过滤掉特殊符号
        filtered = [note for note in generated if note not in ['START', 'END', 'BAR']]
        all_generated.extend(filtered[:12])  # 每个序列取12个音符

    # 可视化生成的音乐
    plt.subplot(1, 2, 2)
    if all_generated:
        note_indices = [tokenizer.note_to_int.get(note, 0) for note in all_generated[:32]]
        plt.plot(note_indices, 'o-', markersize=4)
        plt.title('生成的音乐序列')
        plt.xlabel('时间步')
        plt.ylabel('音符编号')
        plt.grid(True)

    plt.tight_layout()
    plt.show()

    # 生成音频
    if all_generated:
        print(f"\n生成音频 (共{len(all_generated)}个音符)...")
        audio_data, sample_rate = generate_audio_from_notes(all_generated[:24])

        print(f"\n生成的音符序列:")
        print(" -> ".join(all_generated[:24]))

        return audio_data, sample_rate, all_generated[:24]
    else:
        print("没有生成有效的音乐序列")
        return None, None, []

# =====================
# 6. 运行演示
# =====================

if __name__ == "__main__":
    # 运行演示
    result = main_demo()

    if result[0] is not None:
        audio, sr, notes = result
        print(f"\n播放生成的音乐...")
        display(Audio(audio, rate=sr))


    print("完成")
```
